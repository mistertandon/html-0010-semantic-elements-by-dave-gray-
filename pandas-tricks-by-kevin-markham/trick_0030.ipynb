##### Filter a DataFrame by multiple OR conditions

```python
df = pd.read_pickle('../pandas/data_frames/df_dict_of_ridership.pkl', None)
df
```

```python
df_i = df[df.R003.isin([1478, 1560, 1576, 2])]
df_i
```

```python
df_ii = df[df.isin({'R003': [1478, 1560, 1576, 2]})]
df_ii
```

```python
df_iii = df[~df.isin({'R003': [1478, 1560, 1576, 2]})]
df_iii
```
